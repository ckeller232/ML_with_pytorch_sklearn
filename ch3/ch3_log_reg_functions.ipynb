```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid (z):
    return 1.0/(1 + np.exp(-z))
```

```python
z = np.arange(-7,7,0.1)
sigma_z = sigmoid(z)
```

```python
plt.plot(z,sigma_z)
plt.axvline(0.0,color='k')
plt.ylim(-0.1,1)
plt.xlabel('z')
plt.ylabel('$\sigma (z)$')
plt.yticks([0.0,0.5,1.0])
ax=plt.gca()
ax.yaxis.grid(True)
plt.tight_layout()
plt.show()
```

## Loss function

```python
def loss_1(z):
    return -np.log(sigmoid(z))

def loss_0(z):
    return -np.log(1 - sigmoid(z))
```

```python
z = np.arange(-10,10,0.1)
sigma_z = sigmoid(z)
```

```python
c1 = [loss_1(x) for x in z]
c0 = [loss_0(x) for x in z]
```

```python
plt.plot(sigma_z,c1,label='L(w,b) if y=1')
plt.plot(sigma_z,c0,linestyle='--',label='L(w,b) if y=0')
plt.ylim(0.0,5.1)
plt.xlim(0,1)
plt.xlabel('$\sigma(z)$')
plt.ylabel('L(w,b)')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
